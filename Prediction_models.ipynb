```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
fd = pd.read_csv("final_dataset.csv")
```

```python
fd.info()
```

```python
features = [
    'season',
    'year',
    'month',
    'holiday',
    'workingday',
    'weathersit',
    'temp',
    'hum',
    'windspeed'
]

target = 'rentals'
```

```python
x=fd[features]
y=fd[target]
```

```python
cat_features = [
    'season',
    'year',
    'month',
    'holiday',
    'workingday',
    'weathersit'
]

num_features = [
    'temp',
    'hum',
    'windspeed'
]

```

```python
x_encoded=pd.get_dummies(x,columns=cat_features,drop_first=True)
```

```python
x_train , x_test , y_train , y_test=train_test_split(x_encoded,y,test_size=0.2,random_state=42)
```

```python
#Linear regression

lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
y_pred=lr_model.predict(x_test)
```

```python
lr_mae=mean_absolute_error(y_test,y_pred)
```

```python
lr_rmse = mean_squared_error(y_test, y_pred)**0.5

```

```python
lr_r2=r2_score(y_test,y_pred)
```

```python
print("MAE:", lr_mae)
print("RMSE:", lr_rmse)
print("R2 Score:", lr_r2)
```

```python
'''
What this tells us:

Weather + season features are very predictive

Demand follows learnable patterns

Linear model already performs well'''
```

```python
coef_fd = pd.DataFrame({
    "Feature": x_train.columns,
    "Coefficient": lr_model.coef_
})

coef_fd.sort_values(by="Coefficient", ascending=False)
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(x_train, y_train)

y_pred_dt=dt_model.predict(x_test)
```

```python
dt_mae=mean_absolute_error(y_test,y_pred_dt)
dt_mse=mean_squared_error(y_test,y_pred_dt)
dt_rmse=dt_mse**0.5
dt_r2=r2_score(y_test,y_pred_dt)

print("Mean Absolute Error: ", dt_mae)
print("Mean Squared Error: ", dt_mse)
print("Root Mean Squared Error: ", dt_rmse)
print("R2 score: ", dt_r2)
```

```python
#Random Forest
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf_model.fit(x_train, y_train)

y_pred_rf = rf_model.predict(x_test)

rf_mae = mean_absolute_error(y_test, y_pred_rf)
rf_mse = mean_squared_error(y_test, y_pred_rf)
rf_rmse = rf_mse ** 0.5
rf_r2 = r2_score(y_test, y_pred_rf)

print("Random Forest MAE:", rf_mae)
print("Random Forest RMSE:", rf_rmse)
print("Random Forest R2:", rf_r2)
```

```python
'''
🎯 What Happened Conceptually
🔹 Linear Regression

Good baseline
Captures general trends
But assumes linear relationships

🔹 Decision Tree

Too flexible
Likely overfitted
Worse generalization

🔹 Random Forest

Many trees
Averaging reduces overfitting
Captures non-linear patterns
Best balance of bias & variance

Among all tested models, Random Forest performed best with an R² score of 0.87,
 indicating strong predictive capability. It effectively captured non-linear
 relationships between weather and seasonal factors affecting bike rentals.”
 '''
```
