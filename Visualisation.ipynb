```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fd = pd.read_csv("final_dataset.csv")
```

```python
#Visualisation
#Distribution of Bike Rentals

plt.figure(figsize=(10, 6))
plt.hist(fd['rentals'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Rentals')
plt.ylabel('Frequency')
plt.title('Distribution of Rentals')
plt.show()
```

```python
'''
Insight:
1.Peak around average demand
2.Some very high-demand days'''
```

```python
#Rentals vs Temperature
temp_celsius = fd["temp"] * 41
plt.figure(figsize=(8, 5))
plt.scatter(temp_celsius, fd['rentals'], color='green', alpha=0.5)
plt.xlabel('Temperature in Celsius')
plt.ylabel('Rentals')
plt.title('Rentals vs Temperature')
plt.grid(True)
plt.show()
```

```python
'''
Insight:
1.Clear positive relationship
2.Warmer days → more rentals'''
```

```python
#Rentals by Season

Season_name=['Spring', 'Summer', 'Fall', 'Winter']
season_avg = fd.groupby('season')['rentals'].mean()
plt.figure(figsize=(8, 5))
plt.bar(Season_name, season_avg, color='green')
plt.xlabel('Season')
plt.ylabel('Average Rentals')
plt.title('Average Rentals by Season')
plt.show()
```

```python
'''
Insight:
1.Fall season has the highest rentals
2.Spring has the lowest rentals'''
```

```python
#Working vs Non- Working Days
Days=['Non-Working','Working']

plt.figure(figsize=(8, 5))
plt.bar(Days, fd.groupby('workingday')['rentals'].mean(), color='green')
plt.xlabel('Working Day')
plt.ylabel('Average Rentals')
plt.title('Rentals by Working Day')
plt.show()
```

```python
'''
Insight
-More rentals on working days due to office commutes'''
```

```python
fd['weathersit'].value_counts()
```

```python
#Rentals by Weather

plt.figure(figsize=(8, 5))
plt.bar(fd['weathersit'].unique(), fd.groupby('weathersit')['rentals'].mean(), color='green')
plt.xlabel('Weather')
plt.ylabel('Average Rentals')

plt.xticks(
    ticks=[1,2,3],
    labels=['Clear / Few clouds / Partly cloudy',
            'Mist + Cloudy',
            'Light Snow / Light Rain + Thunderstorm'],
    rotation=45
)
plt.title('Rentals by Weather')
plt.show()
```

```python
'''
Insight:

Cloudy → highest rentals

 rain → lowest'''
```

```python
#Rentals by month

plt.figure(figsize=(8, 5))
plt.bar(fd.groupby('month')['rentals'].unique().index, fd.groupby('month')['rentals'].mean(), color='green', alpha=0.5)
plt.xlabel('Month')
plt.ylabel('Rentals')
plt.title('Rentals by Month')
month_names = ['Jan','Feb','Mar','Apr','May','Jun',
               'Jul','Aug','Sep','Oct','Nov','Dec']

plt.xticks(
    fd.groupby('month')['rentals'].mean().index,
    month_names
)

plt.grid(True)
```

```python
'''
Insights:
-Summer has the highest rentals
-Winter has the lowest rentals'''
```
