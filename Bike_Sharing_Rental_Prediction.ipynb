```python
df_clean = df.copy()

df_clean.drop(
    columns=["instant","dteday","casual","registered","atemp"],
    errors="ignore",
    inplace=True
)
```

```python
df_clean.duplicated().sum()
```

```python
categorical_cols = [
    "season", "yr", "mnth",
    "holiday", "workingday", "weathersit"
]
categorical_cols = ["season","year","month","holiday","workingday","weathersit"]

existing_cols = [c for c in categorical_cols if c in df_clean.columns]

df_clean[existing_cols] = df_clean[existing_cols].astype("category")
```

```python
df_clean.rename(columns={"cnt":"rentals","mnth":"month","yr":"year"},inplace=True)
```
