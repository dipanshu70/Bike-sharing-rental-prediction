```
df_clean=df.copy()
```

```
df_clean.drop(columns=["instant", "dteday", "casual", "registered", "atemp"],inplace=True)
```

```
df_clean.duplicated().sum()
```

```
categorical_cols = [
    "season", "yr", "mnth",
    "holiday", "workingday", "weathersit"
]

df_clean[categorical_cols] = df_clean[categorical_cols].astype("category")
```

```
df_clean.rename(columns={"cnt":"rentals","mnth":"month","yr":"year"},inplace=True)
```
